```python
# Import the Pandas library
import pandas as pd
```

```python
# Create a reference the CSV file desired

csv_file_path = './Resources/ufoSightings.csv'

# Read the CSV into a Pandas DataFrame
df = pd.read_csv(csv_file_path)

# Print the first five rows of data to the screen
df.head()
```

```python
# Check to see if there are any rows with missing data
df.count()
```

```python
# Remove the rows with missing data
df_cleaned = df.dropna(how='any')
df_cleaned.count()
```

```python
# Filter the data so that only those sightings in the US are in a DataFrame
us_sightings = df_cleaned.loc[df_cleaned['country']=='us',:]
us_sightings.head()
```

```python
# Count how many sightings have occured within each state
state_counts = df_cleaned['state'].value_counts()
state_counts
```

```python
# Convert the state_counts Series into a DataFrame
df_state_counts = pd.DataFrame(state_counts)
df_state_counts.head()
```

```python
# Convert the column name into "Sum of Sightings"
df_state_counts = df_state_counts.rename(columns={"state": "Sum of Sightings"})
df_state_counts.index.name = 'State'
df_state_counts.head()
```

```python
# Want to add up the seconds UFOs are seen? There is a problem
# Problem can be seen by examining datatypes within the DataFrame
df_cleaned['duration (seconds)'] = pd.to_numeric(df_cleaned['duration (seconds)'])
df_cleaned.dtypes
```
